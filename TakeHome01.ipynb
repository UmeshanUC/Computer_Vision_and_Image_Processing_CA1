```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def AdjustImageIntensityLevels(imageMat, noOfLevels):
    adjustedImg = imageMat/255*(noOfLevels)
    return np.round(adjustedImg)


def PlotImagesInRow(img, newImage):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
    ax1.imshow(img)
    ax2.imshow(newImage)
    plt.gray()
    plt.show()
```

#### Q1.

```python
imgRaw = cv2.imread('assets/astronaut.jpg', 0 )
img = cv2.resize(imgRaw,(0,0), fx=0.5, fy=0.5)
```

```python
intensityLevels = int(input('intensityLevels as integer:'))
newImage1 = AdjustImageIntensityLevels(img, intensityLevels)

PlotImagesInRow(img, newImage1)
```

#### Q2.

##### 3x3 average filter

```python
arr1 = np.ones((3,3),np.float32)/9
newImage2 = cv2.filter2D(imgRaw,-1,arr1)

PlotImagesInRow(img, newImage2)
```

##### 10x10 average filter

```python
arr2 = np.ones((10,10),np.float32)/100
newImage3 = cv2.filter2D(img,-1,arr2)

PlotImagesInRow(img, newImage3)
```

##### 20x20 average filter

```python
arr3 = np.ones((20,20),np.float32)/400
newImage4 = cv2.filter2D(img,-1,arr3)

PlotImagesInRow(img, newImage4)
```

#### Q3.

```python
# getting rows and cols in img matrix
(rows1,cols1) = img.shape[:2]

# 45 deg
M1 = cv2.getRotationMatrix2D((cols1/2,rows1/2), 45, 1)
# 90 deg
M2 = cv2.getRotationMatrix2D((cols1/2,rows1/2), 90, 1)

# apply 45deg to img
newImage5 = cv2.warpAffine(img, M1, (cols1,rows1))
# apply 45deg to img
newImage6 = cv2.warpAffine(img, M2, (cols1,rows1))

# plot 45deg adjusted image
PlotImagesInRow(img, newImage5)

# plot 90deg adjusted image
PlotImagesInRow(img, newImage6)
```

#### Q4.

```python
# for 3x3 block
arr4 = np.ones((3,3),np.float32)/9
newImage7 = cv2.filter2D(img,-1,arr4)

# for 5x5 block
arr5 = np.ones((5,5),np.float32)/25
newImage8 = cv2.filter2D(img,-1,arr5)

# for 7x7 block
arr6 = np.ones((7,7),np.float32)/49
newImage9 = cv2.filter2D(img,-1,arr6)

# Save the new images
PlotImagesInRow(img, newImage7)
PlotImagesInRow(img, newImage8)
PlotImagesInRow(img, newImage9)
```
